```python
import pandas as pd
dados = pd.read_csv('data/car-prices.csv')
dados.head()
```

```python
# Traduzindo colunas
mapa = {
    'mileage_per_year': 'milhas_por_ano',
    'model_year': 'ano_modelo',
    'price': 'preco',
    'sold': 'vendido'
}
dados = dados.rename(columns = mapa)
dados.head()
```

```python
# 0 => nÃ£o vendido, 1 => vendido
a_trocar = {
    'no': 0,
    'yes': 1
}
dados.vendido = dados.vendido.map(a_trocar)
dados.head()
```

```python
# Criando coluna da idade do carro
from datetime import datetime

ano_atual = datetime.today().year
dados['idade_modelo'] = ano_atual - dados.ano_modelo
dados.head()
```

```python
# Transforma milhas em quilometros
dados['km_por_ano'] = dados.milhas_por_ano * 1.60934
dados.head()
```

```python
# Deleta as colunas que nÃ£o serÃ£o utilizadas no modelo
dados.drop(columns = ['Unnamed: 0', 'milhas_por_ano', 'ano_modelo'], axis = 1, inplace = True)
dados.head()
```

```python
# Dados de entrada (x) x Dados de saida (y)
x = dados[['preco', 'idade_modelo', 'km_por_ano']]
y = dados['vendido']
```

```python
# Testando modelo linear
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

seed = 20

treino_x, teste_x, treino_y, teste_y = train_test_split(x,y, test_size = 0.25, random_state = seed, stratify = y)

print("Treinaremos com {} elementos e testaremos com {} elementos".format(len(treino_x), len(teste_x)))

modelo = LinearSVC(random_state = seed)
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100

print("A acuracia foi: ", round(acuracia,2), '%')
```

```python
# Comparando acuracia com um Dummy
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(random_state = seed)
dummy.fit(treino_x, treino_y)
previsoes = dummy.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100

print("A acuracia do dummy foi: ", round(acuracia,2), '%')
```

```python
# Aplicando modelo nÃ£o linear e com escala padronizada
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

seed = 20

raw_treino_x, raw_teste_x, treino_y, teste_y = train_test_split(x,y, test_size = 0.25, random_state = seed, stratify = y)

print("Treinaremos com {} elementos e testaremos com {} elementos".format(len(treino_x), len(teste_x)))

scaler = StandardScaler()
scaler.fit(raw_treino_x)
treino_x = scaler.transform(raw_treino_x)
teste_x = scaler.transform(raw_teste_x)

modelo = SVC(random_state = seed)
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100

print("A acuracia foi: ", round(acuracia,2), '%')
```

```python
# Aplicando modelo nÃ£o linear e com escala padronizada
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

seed = 20

raw_treino_x, raw_teste_x, treino_y, teste_y = train_test_split(x,y, test_size = 0.25, random_state = seed, stratify = y)

print("Treinaremos com {} elementos e testaremos com {} elementos".format(len(treino_x), len(teste_x)))

modelo = DecisionTreeClassifier(random_state = seed, max_depth=3)
modelo.fit(raw_treino_x, treino_y)
previsoes = modelo.predict(raw_teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100

print("A acuracia foi: ", round(acuracia,2), '%')
```

```python
from sklearn.tree import export_graphviz
import graphviz

features = x.columns
dot_data = export_graphviz(modelo, feature_names = features, 
                           filled = True, rounded = True,
                           class_names = ['nÃ£o', 'sim'])
grafico = graphviz.Source(dot_data)
grafico
```
