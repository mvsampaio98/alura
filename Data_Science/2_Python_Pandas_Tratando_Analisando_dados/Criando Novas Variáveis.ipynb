## Criando Novas Variaveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head()
```

```python
dados['Valor_Bruto'] = dados['Valor'] + dados['Condominio'] + dados ['IPTU']
dados['Valor_Bruto']
```

```python
dados
```

```python
dados['Valor_m2'] = dados['Valor']/dados['Area']
dados['Valor_m2'] = dados['Valor_m2'].round(2)
dados
```

```python
dados['Valor_Bruto_m2'] = dados['Valor_Bruto']/dados['Area']
dados['Valor_Bruto_m2'] = dados['Valor_Bruto_m2'].round(2)
dados
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

## Excluindo Variáveis

```python
dados_aux = pd.DataFrame(dados[['Valor_Bruto', 'Valor_m2','Valor_Bruto_m2', 'Tipo Agregado']])
```

```python
dados_aux
```

```python
del dados_aux['Valor_Bruto']
dados_aux
```

```python
dados_aux.pop('Valor_Bruto_m2')
```

```python
dados_aux
```

```python
dados.drop(['Valor_Bruto', 'Valor_Bruto_m2'], axis= 1, inplace = True)
```
