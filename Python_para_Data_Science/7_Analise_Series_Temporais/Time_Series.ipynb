# Aula 2

```python
# Criar colunas de diferença entre as vendas de um mês para outro
alucar['aumento'] = alucar['vendas'].diff()
alucar.head()
```

```python
alucar_vendas = sns.set_palette('Accent')
alucar_vendas = sns.set_style('darkgrid')
alucar_vendas = sns.lineplot(x='mes', y='aumento', data=alucar)
alucar_vendas.figure.set_size_inches(12,8)
alucar_vendas.set_title('Aumento das vendas da Alucar de 2017 e 2018', loc = 'left', fontsize=18)
alucar_vendas.set_xlabel('Tempo',fontsize=14)
alucar_vendas.set_ylabel('Aumento',fontsize=14)
alucar_vendas = alucar_vendas
```

```python
def plotar(titulo, labelx, labely, x, y, dataset):
    alucar_vendas = sns.set_palette('Accent')
    alucar_vendas = sns.set_style('darkgrid')
    alucar_vendas = sns.lineplot(x = x, y = y, data = dataset)
    alucar_vendas.figure.set_size_inches(12,8)
    alucar_vendas.set_title(titulo, loc = 'left', fontsize=18)
    alucar_vendas.set_xlabel(labelx,fontsize=14)
    alucar_vendas.set_ylabel(labely,fontsize=14)
    alucar_vendas = alucar_vendas
```

```python
plotar('Aumento das vendas da Alucar de 2017 e 2018', 'Tempo', 'Aumento',
      'mes', 'aumento', alucar)
```

```python
alucar['aceleracao'] = alucar['aumento'].diff()
```

```python
alucar.head()
```

```python
plotar('Aceleração das vendas da Alucar de 2017 e 2018', 'Tempo', 'Aceleração',
       'mes', 'aceleracao', alucar)
```

```python
plt.figure(figsize=(16,12))
alucar_vendas = plt.subplot(3,1,1)
alucar_vendas.set_title('Análise de vendas da Alucar de 2017 e 2018', fontsize=18,loc='left')
sns.lineplot(x='mes',y='vendas',data=alucar)
plt.subplot(3,1,2)
sns.lineplot(x='mes',y='aumento',data=alucar)
plt.subplot(3,1,3)
sns.lineplot(x='mes',y='aceleracao',data=alucar)
alucar_vendas = alucar_vendas
```

```python
def plotar_comparacao(x, y1, y2, y3, dataset, titulo):
    plt.figure(figsize=(16,12))
    alucar_vendas = plt.subplot(3,1,1)
    alucar_vendas.set_title(titulo, fontsize=18,loc='left')
    sns.lineplot(x=x,y=y1,data=dataset)
    plt.subplot(3,1,2)
    sns.lineplot(x=x,y=y2,data=dataset)
    plt.subplot(3,1,3)
    sns.lineplot(x=x,y=y3,data=dataset)
    alucar_vendas = alucar_vendas
```

```python
plotar_comparacao('mes','vendas','aumento','aceleracao',alucar,
                  'Análise das vendas da Alucar de 2017 e 2018')
```

```python
from pandas.plotting import autocorrelation_plot
```

```python
alucar_vendas = plt.figure(figsize=(12,6))
alucar_vendas.suptitle('Correlação das vendas', fontsize=18, x = 0.26, y = 0.95)
alucar_vendas = autocorrelation_plot(alucar['vendas'])
```

```python
alucar_vendas = plt.figure(figsize=(12,6))
alucar_vendas.suptitle('Correlação do aumento', fontsize=18, x = 0.26, y = 0.95)
alucar_vendas = autocorrelation_plot(alucar['aumento'][1:])
```

```python
alucar_vendas = plt.figure(figsize=(12,6))
alucar_vendas.suptitle('Correlação da aceleração', fontsize=18, x = 0.26, y = 0.95)
alucar_vendas = autocorrelation_plot(alucar['aceleracao'][2:])
```

**Análisando a Sazonalidade**

```python
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia'].dt.day_name()
```

```python
vendas_por_dia.head()
```

```python
vendas_por_dia['dia_da_semana'].unique()
```

```python
dias_traduzidos = {'Monday': 'Segunda', 'Tuesday': 'Terca', 'Wednesday': 'Quarta',
                   'Thursday': 'Quinta', 'Friday': 'Sexta', 'Saturday': 'Sabado',
       'Sunday': 'Domingo'}
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia_da_semana'].map(dias_traduzidos)
vendas_por_dia.head(15)
```

**Agrupando os dias**

```python
vendas_agrupadas = vendas_por_dia.groupby('dia_da_semana')['vendas','aumento','aceleracao'].mean().round()
```

```python
vendas_agrupadas
```

## Statsmodels

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado = seasonal_decompose(chocolura['vendas'], period=3)
ax = resultado.plot()
```

```python
observacao = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```python
data = ({
    'observacao':observacao,
    'tendencia':tendencia,
    'sazonalidade':sazonalidade,
    'ruido':ruido
})
resultado = pd.DataFrame(data)
resultado.head()
```

```python
plotar_comparacao(resultado.index, 'observacao', 'tendencia','sazonalidade',
                 resultado, 'Exemplo de Statesmodels')
```
